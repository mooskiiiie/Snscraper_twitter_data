```python
import snscrape.modules.twitter as sntwitter
import pandas as pd
```

```python
# Creating list to append tweet data
tweets_list = []

# Using TwitterSearchScraper to scrape data and append tweets to list
for i, tweet in enumerate(sntwitter.TwitterSearchScraper('tallano gold since:2021-10-01 until:2022-1-25').get_items()):
    if i>10000:
        break
    tweets_list.append([tweet.date, tweet.url, tweet.id, tweet.content, tweet.user.username, 
                         tweet.user.description, tweet.user.verified, tweet.retweetedTweet, tweet.quotedTweet, tweet.mentionedUsers])
    
# Creating a dataframe from the tweets list above
tweets_df = pd.DataFrame(tweets_list, columns=['Datetime', 'URL', 'Tweet Id', 'Text', 'Username', 'Bio', 'Verified', 'Retweeted', 'Quoted', 'Mentioned Users'])
```
